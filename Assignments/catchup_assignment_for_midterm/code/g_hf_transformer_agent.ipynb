```python
# import libraries
from transformers import HfAgent
```

```python
# configure
agent = HfAgent("https://api-inference.huggingface.co/models/bigcode/starcoder2-15b")
```

```python
# run a command
prompt = "Generate a plot of a sine wave using Python and matplotlib."
result = agent.run(prompt)
```

```python
# show results
print("\n Agent Output:\n")
print(result)
```
