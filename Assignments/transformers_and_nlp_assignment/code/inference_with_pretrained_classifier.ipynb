```python
# import necessary modules
import tensorflow as tf
import keras_nlp

# initialize preprocessing and model components for BERT
bert_tokenizer = keras_nlp.models.BertPreprocessor.from_preset("bert_base_en_uncased")
sentiment_model = keras_nlp.models.BertClassifier.from_preset(
    "bert_base_en_uncased",
    num_classes=2  # binary sentiment classification
)
```

```python
# sample input for sentiment analysis
review_text = "I absolutely loved the performance and storyline. A must-watch!"

# tokenize input text using BERT tokenizer
encoded_input = bert_tokenizer(tf.constant([review_text]))
```

```python
# perform classification
raw_output = sentiment_model(encoded_input)
probabilities = tf.nn.softmax(raw_output)

# interpret the prediction
class_names = ["Negative", "Positive"]
predicted_index = tf.argmax(probabilities[0]).numpy()
predicted_sentiment = class_names[predicted_index]
predicted_confidence = tf.reduce_max(probabilities[0]).numpy()
```

```python
# results
print(f"Review: {review_text}")
print(f"Predicted Sentiment: {predicted_sentiment} (Confidence: {predicted_confidence:.2f})")
```
